```python
import numpy as np
import pandas as pd
import matplotlib.pyplot
import sklearn
```

```python
df = pd.read_csv("data/train.csv")
df
```

```python
a = df["Cabin"][0]
df.info()
df["no_cabin"] = (df["Cabin"].isnull())


from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(["True","False"])
df["no_cabin"] = label.transform(df["no_cabin"])

label = LabelEncoder()
label.fit(["male","female"])
df["Sex"] = label.transform(df["Sex"])

```

```python
df = df.join(pd.get_dummies(df["Cabin"],dtype=int))
```

```python
df = df[df.columns.drop(["PassengerId","Name","SibSp","Ticket","Fare","Cabin","Embarked"])]
df = df.dropna()
```

```python

df["Age"]=df["Age"].astype(int)
df.dtypes
```

```python
df 
```

```python
x = df[df.columns.drop(["Survived"])]
y = df["Survived"]

from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(C = 0.5,kernel="linear")
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report
yhat = model.predict(xtest)
print(classification_report(ytest,yhat))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver= "saga",C = 0.5)
model.fit(xtrain,ytrain)
```

```python

yhat = model.predict(xtest)
print(classification_report(ytest,yhat))
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = "gini")
model.fit(xtrain,ytrain)
```

```python
yhat = model.predict(xtest)
print(classification_report(ytest,yhat))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(xtrain,ytrain)
```

```python
yhat = model.predict(xtest)
print(classification_report(ytest,yhat))
```
